```python
from bs4 import BeautifulSoup
import requests as re
import pandas as pd
```

```python
webpage=re.get("https://stats.espncricinfo.com/ci/content/records/83548.html")
print(webpage)
```

```python

record=BeautifulSoup(webpage.content)
table=record.find_all("table")
list=[]
```

```python
for i in table:
    for j in i.find_all("td"):
        list.append(j.text.replace("\n", " "))
print(list)
Player_Name=list[0:len(list):15]
Player_Name1=Player_Name[0:92]
Career_Span=list[1:len(list):15]
Career_Span1=Career_Span[0:92]
Matches=list[2:len(list):15]
Matches1=Matches[0:92]
Innings=list[3:len(list):15]
Innings1=Innings[0:92]
Not_Out=list[4:len(list):15]
Not_Out1=Not_Out[0:92]
Highest=list[5:len(list):15]
Highest1=Highest[0:92]
Average=list[7:len(list):15]
Average1=Average[0:92]
Ball_Faced=list[8:len(list):15]
Ball_Faced1=Ball_Faced[0:92]
Strike_Rate=list[9:len(list):15]
Strike_Rate1=Strike_Rate[0:92]
Century=list[10:len(list):15]
Century1=Century[0:92]
Half_Century=list[11:len(list):15]
Half_Century1=Half_Century[0:92]
Ducks=list[12:len(list):15]
Ducks1=Ducks[0:92]
Fours=list[13:len(list):15]
Fours1=Fours[0:92]
Sixes=list[14:len(list):15]
Sixes1=Sixes[0:92]
```

```python

df=pd.DataFrame({})
df["Player_Name"]=Player_Name1
df["Career_Span"]=Career_Span1
df["Matches"]=Matches1
df["Innings"]=Innings1
df["Not_Out"]=Not_Out1
df["Highest Score"]=Highest1
df["Average"]=Average1
df["Balls_Faced"]=Ball_Faced1
df["Strike_Rate"]=Strike_Rate1
df["Century"]=Century1
df["Half_Century"]=Half_Century1
df["Ducks"]=Ducks1
df["Fours"]=Fours1
df["Sixes"]=Sixes1
print(df)
```
